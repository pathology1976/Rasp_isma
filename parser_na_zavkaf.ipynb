```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def parse_igma_kaf(url):
    response = requests.get(url)

    if response.status_code == 200: # 200 - это ответ сайта, непонятная херня типа 404
        soup = BeautifulSoup(response.text, 'html.parser')

        #Ищет строки таблицы. tbody в html - это таблица короче
        rows = soup.select('tbody tr:not(:first-child)')

        departments_data = []
        for row in rows:
            #беру из тбоди 1 и 2 ячейку и прогоняю по циклу все tr из tbody
            columns = row.find_all('td') # td - тоже какая-то хрень из разметки
            department_name = columns[1].text.strip()
            head_name = columns[2].text.strip()

            departments_data.append({
                'Кафедра': department_name,
                'Заведующий': head_name
            })
        df = pd.DataFrame(departments_data)
        return df
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

url = "https://www.igma.ru/kaf"
df = parse_igma_kaf(url)
print(df)
df.to_csv("zavcaf1.csv")
```
